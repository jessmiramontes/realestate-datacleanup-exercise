#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')
valores_deseados = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alorcón']
new_data = ds[ds['level5'].isin(valores_deseados)]

#Para normalizar los precios de Fuenlabrada
fuenlabrada_data = new_data[new_data['level5'] == 'Fuenlabrada']
xfuen = fuenlabrada_data['price']
normalized_pricefuen = (xfuen - xfuen.min()) / (xfuen.max() - xfuen.min())

#Para normalizar los precios de Leganés
leganes_data = new_data[new_data['level5'] == 'Leganés']
xleg = leganes_data['price']
normalized_priceleg = (xleg - xleg.min()) / (xleg.max() - xleg.min())

#Para normalizar los precios de Getafe
getafe_data = new_data[new_data['level5'] == 'Getafe']
xget = getafe_data['price']
normalized_priceget = (xget - xget.min()) / (xget.max() - xget.min())

#Para normalizar los precios de Alcorcón
alorcon_data = new_data[new_data['level5'] == 'Alcorcón']
xal = alorcon_data['price']
normalized_priceal = (xal - xal.min()) / (xal.max() - xal.min())

# Generar el histograma
plt.figure(figsize=(10, 6))
plt.hist(normalized_pricefuen, alpha=0.5, label='Fuenlabrada')
plt.hist(normalized_priceleg, alpha=0.5, label='Leganés')
plt.hist(normalized_priceget, alpha=0.5, label='Getafe')
plt.hist(normalized_priceal, alpha=0.5, label='Alcorcón')

plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.title('Histograma de Precios Normalizados por Localidad')
plt.legend(loc='upper right')
plt.show()

```

En general, parece que hay una tendencia a que la mayoría de las viviendas en estas localidades tengan precios más bajos, con algunas variaciones entre las localidades. Getafe y Fuenlabrada tienen concentraciones claras en precios relativamente bajos, mientras que Leganés y Alcorcón muestran más variabilidad.

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in Exercise 14 (★★☆)

```python
## HERE: plot the coordinates of the estates

## PUT HERE YOUR CODE:
from ipyleaflet import Marker, CircleMarker
import pandas as pd
import matplotlib.pyplot as plt


# 1. Cargar los datos
ds = pd.read_csv('assets/real_estate.csv', sep=';')
ds['latitude'] = ds['latitude'].astype("string")
ds['longitude'] = ds['longitude'].astype("string")
ds['latitude'] = ds['latitude'].replace(',', '.', regex=True)
ds['longitude'] = ds['longitude'].replace(',', '.', regex=True)

# 2. Filtrar los datos por las poblaciones deseadas
valores_deseados = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']
filtered_data = ds[ds['level5'].isin(valores_deseados)]

# 3. Crear un diccionario con las coordenadas
coordinates_dict = {}

for location in valores_deseados:
    location_data = filtered_data[filtered_data['level5'] == location]
    coordinates_dict[location] = list(zip(location_data['latitude'], location_data['longitude']))

# 4. Graficar las coordenadas por color de cada población
plt.figure(figsize=(10, 8))

colors = ['red', 'blue', 'green', 'orange']  # Colores para cada población
for i, location in enumerate(valores_deseados):
    coords = coordinates_dict[location]
    
    for items in coords:
        lats, lons = zip(items)  # Desempaquetar latitudes y longitudes
        mkers = CircleMarker(location=(lats[0], lons[0]), radius=6, color=colors[i], fill=True, fill_color=colors[i], fill_opacity=1)
        map.add_layer(mkers)
   # plt.scatter(lons, lats, color=colors[i], label=location, alpha=0.6, s=10)

map
```
